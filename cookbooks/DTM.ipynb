This notebooks shows code snippets to generate dynamic topographic map for multiple protein snapshots or a trajectory.

```python
import alphaspace2 as al
import mdtraj as md
import numpy as np
```

```python
receptors_path = ["/Users/haotian/Downloads/DTM/bcl2/prod_{}.pdb".format(i) for i in range(1,10)]
```

```python
al_ss_list = []
for receptor_path in receptors_path:
    receptor = md.load(receptor_path)
    ss = al.Snapshot()
    ss.run(receptor = receptor)
    al_ss_list.append(ss)
```

```python
trajectory = al.Trajectory(al_ss_list)
```

```python
al.View.write_trajectory(trajectory=trajectory,folder_path="/Users/haotian/Downloads/DTM/bcl2")
```
