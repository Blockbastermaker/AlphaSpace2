Bin Clustering Example

Here is an example of using the bin clustering to cluster large number of beta atoms into pockets. 

To use the iterload function, you 

```python
import alphaspace
import mdtraj as md
import nglview as nv
import numpy as np
```

```python
universe = alphaspace.load_from_file(path_file='2mwy')
```

```python
beta_xyz = universe.beta_xyz()
bin_size = [1.8, 1.8, 1.8]
beta_pocket_labels = alphaspace.AS_Funct.bin_cluster(beta_xyz * 10, bin_size=bin_size,distance = 4.7)
# beta_xyz * 10 is because mdtraj use nm insead of Angstrom, this converts it to A

for beta_idx, pocket_idx in enumerate(beta_pocket_labels):
    snapshot_idx, snapshot_beta_idx = universe.map_beta(beta_idx)
    probe_scores = universe[snapshot_idx].beta_scores[snapshot_beta_idx]

    best_probe_element = alphaspace.AS_Vina.probe_element_list[np.argmin(probe_scores)]
    best_probe_score = probe_scores[np.argmin(probe_scores)]

    print("Beta Atom {} in snapshot {} belongs to pocket {} with {} of score {}".format(snapshot_beta_idx,
                                                                                        snapshot_idx, pocket_idx,
                                                                                        best_probe_element,
                                                                                        best_probe_score))
```
