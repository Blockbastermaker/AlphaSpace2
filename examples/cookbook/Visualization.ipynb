```python
# This section shows how you can show D-pocket information.
import alphaspace as al
import mdtraj
import numpy as np

traj = mdtraj.load_hdf5("/Users/haotian/Desktop/bcl2.holo1.10.h5",stride = 100)

traj.superpose(traj,0,atom_indices=[atom.index for atom in traj.top.atoms if not atom.residue.is_water],parallel=True)

u = al.AS_Universe(traj,guess_receptor_binder=True,guess_by_order=False)
```

```python
u.run_alphaspace_mp()
```

```python
view = u.draw(snapshot_idx=0,show_binder=True,surface_opacity=0)
view
```

```python
d_pockets = u._gen_d_pockets()
```

```python
d_pockets = u._gen_d_pockets_iter(sample_frames=40, sample_ratio=5, pocket_space_cutoff=20)
```

```python
stability = [len(set([pocket.snapshot_idx for pocket in i])) for i in d_pockets.values()]
spaces = [np.average([pocket.space for pocket in i]) for i in d_pockets.values()]
centers = [list(np.average([pocket.centroid for pocket in i],axis = 0)*10) for i in d_pockets.values()]
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(d_pockets)),sorted(stability))
plt.show()
```

```python
def rgb(minimum, maximum, value):
    minimum, maximum = float(minimum), float(maximum)
    ratio = 2 * (value - minimum) / (maximum - minimum)
    b = int(max(0, 255 * (1 - ratio)))
    r = int(max(0, 255 * (ratio - 1)))
    g = 255 - b - r
    return [r, g, b]
for i in range(len(stability)):
    u._draw_as_sphere(centers[i],color= rgb(min(spaces),max(spaces),spaces[i]),
                      radius= stability[i] / 1000 + 0.5,opacity=1)
```

```python
scores = []
for dp in u.d_pockets:
    scores.append(-np.average([pocket.score for pocket in dp]))

second_best = sorted(scores,reverse=True)[int(len(scores)/10)]

for i,dp in enumerate(u.d_pockets):
    centroid = list(np.average([pocket.centroid for pocket in dp],axis=0)* 10)
    stability  = len(set([pocket.snapshot_idx for pocket in dp]))
    score = scores[i]
    u._draw_as_sphere(centroid,color= rgb(min(scores),second_best,score),
                      radius= stability / 10 + 0.5,opacity=0.9)
#     print(np.average([pocket.score for pocket in dp]))
#     u._draw_as_sphere(centroid,color=rgb(1,160,160),radius=-np.average([pocket.score for pocket in dp]),opacity=len(set([pocket.snapshot_idx for pocket in dp])) / 320 + 0.5)
```

```python
union_overlap, time_overlap = u._connect_d_pockets()
overlap = union_overlap - time_overlap
```

```python
from itertools import combinations
for i, j in combinations(range(len(d_pockets)),2):
    if overlap[i,j] > 10:
        print('hit')
        position1 = list(np.average([pocket.centroid for pocket in d_pockets[i]],axis=0)* 10)
        position2 = list(np.average([pocket.centroid for pocket in d_pockets[j]],axis=0)* 10)
        u._draw_cylinder(position1=position1,position2 = position2, color=[0,0,0],radius=[overlap[i,j] / 100],opacity=1)
    
```
