```python
# This section shows how you can show D-pocket information.
import alphaspace as al
import mdtraj
import numpy as np

traj = mdtraj.load_hdf5("/Users/haotian/Desktop/bcl2.holo1.10.h5",stride = 10)

traj.superpose(traj,0,atom_indices=[atom.index for atom in traj.top.atoms if not atom.residue.is_water],parallel=True)

u = al.AS_Universe(traj,guess_receptor_binder=True,guess_by_order=False)
```

```python
u.run_alphaspace_mp()
```

```python
view = u.draw(snapshot_idx=0,show_binder=True,surface_opacity=0)
view
```

```python
for i in u.snapshots_indices:

    pocket_network = u._connect_pockets(i)
    for p1,p2 in pocket_network.edges:


        position1, position2 = list(p1.centroid* 10),list(p2.centroid * 10)
        print(position1,position2)

        u._draw_cylinder(position1=position1,position2 = position2, color=[0,0,0],radius= [0.1] ,opacity=0.8)
    
```
