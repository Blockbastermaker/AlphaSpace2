```python
import alphaspace
import nglview as nv
import mdtraj as md

lig_path = '3MXF_A_lig.pdb'
prot_path = '3MXF_A_rep.pdbqt'

prot = alphaspace.AS_IO.load_pdbqt(prot_path)

universe = alphaspace._load(files=prot_path)
universe.screen_by_contact(md.load(lig_path).xyz,cutoff=3.5)
```

```python
view = nv.show_mdtraj(prot)
view.add_trajectory(md.load(lig_path))
# view.add_representation('surface')
view
```

```python
universe.draw_pocket(view,contact_only=True,radius=1,snapshot_idx=0)
```

```python
for pocket in universe.snapshots[0].pockets:
    if pocket.is_contact:
        print(pocket.index,pocket.space,  pocket.score)
```
